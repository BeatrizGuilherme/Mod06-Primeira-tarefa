```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sinasc = pd.read_csv('SINASC_RO_2019-Copy1.csv')
print(sinasc.shape)
```

```python
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
sinasc.columns
```

```python
variaveis= sinasc[['APGAR1','APGAR5','SEXO','GESTACAO','ESTCIVMAE']]
variaveis
```

```python
variaveis.columns
```

```python
plt.rc('figure', figsize=(10, 8))
fig, eixos = plt.subplots(3, 2)

sns.countplot(ax = eixos[0, 0], x='APGAR1', data=variaveis)

sns.countplot(ax = eixos[0, 1], x='APGAR5', data=variaveis)

sns.countplot(ax = eixos[1, 0], x='SEXO', data=variaveis)

sns.countplot(ax = eixos[1, 1], x='GESTACAO', data=variaveis)

sns.countplot(ax = eixos[2, 0], x='ESTCIVMAE', data=variaveis)
```

```python
plt.close('all')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
variaveis_cons= sinasc[['SEXO','GESTACAO','ESTCIVMAE','CONSULTAS']]
variaveis_cons
```

```python
plt.rc('figure', figsize=(10, 8))
fig, eixos = plt.subplots(2, 2)

sns.countplot(ax = eixos[0, 0], x='SEXO', data=variaveis_cons)

sns.countplot(ax = eixos[0, 1], x='GESTACAO', data=variaveis_cons)

sns.countplot(ax = eixos[1, 0], x='ESTCIVMAE', data=variaveis_cons)

sns.countplot(ax = eixos[1, 1], x='CONSULTAS', data=variaveis_cons)

```

```python
plt.close('all')
```
